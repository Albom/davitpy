```python
from pylab import *
from numpy import outer
rc('text', usetex=False)
a=outer(ones(10),arange(0,1,.1))
figure(figsize(26,30))
subplots_adjust(top=0.8,bottom=0.05,left=.2,right=0.99,wspace=.05)
maps=[m for m in cm.datad if not m.endswith("_r")]
maps.sort()
l=len(maps)+1
for i, m in enumerate(maps):
    subplot(l,3,i+1)
    ax = gca()
    axis("off")
    imshow(a,aspect='auto',cmap=get_cmap(m),origin="lower")
    ax.annotate(m, xy=(0,3), fontsize=10, bbox=dict(boxstyle="round", fc="w"))
```
