```python
import datetime as dt
import datetime
from models import raydarn
import pandas
import numpy as np
import os
```

```python
dict_list = []
for kt in rto.scatter.isc.keys():
    for kb in rto.scatter.isc[kt].keys():
        elv_dict = rto.scatter.isc[kt][kb]
        for key,value in elv_dict.items():
            for ii in range(value['nstp']):
                tmp_dict = {'time':kt,'beam':kb,'elv_init':key,'radar':radar}
                for param,param_val in value.items():
                    if param == 'nstp': continue
                    tmp_dict[param] = param_val[ii]
                dict_list.append(tmp_dict)
```

```python
df = pandas.DataFrame(dict_list)
```

```python
dict_list_gs = []
for kt in rto.scatter.gsc.keys():
    for kb in rto.scatter.gsc[kt].keys():
        elv_dict_gs = rto.scatter.gsc[kt][kb]
        for key,value in elv_dict_gs.items():
            tmp_dict_gs = {'time':kt, 'beam':kb, 'elv_init':key, 'radar':radar}
            dict_list_gs.append(tmp_dict_gs)
                             
```

```python
df['UTtime'] = np.zeros_like(df['lon'], dtype = np.float)
df['Loc_time'] = np.zeros_like(df['lon'], dtype = np.float)
size=df['time'].size 
for index in range(0, size-1):
    time_shift = df['lon'][index]/15.
    hysteresis = 0.    #was 1.
    UThour = df['time'][index].hour
    UTminute = df['time'][index].minute
    df['UTtime'][index] = UThour + UTminute/60.
    local_time = df['UTtime'][index] + (time_shift - hysteresis)
    if local_time < 0:
        local_time = local_time + 24.
    elif local_time > 24.:
        local_time = local_time - 24.
    df['Loc_time'][index] = local_time 
df['UTtime'][size-1] = UThour +UTminute/60.
df['Loc_time'][size-1] =local_time
    
    
```

```python
df['h_2p'] = np.sqrt(6370.**2 + (df['gran']*1e-3)**2 + 2*(df['gran']*1e-3)*6370.*np.sin(df['elv_init']*np.pi/180.)) - 6370.
df['h_2p'].describe()
```

```python
tmp = []
for index,row in df.iterrows():
    h_2p_max = 400. + 75*np.cos(row['Loc_time']*np.pi/24.)  #use 2000 for pure 2p-method, 550 for modified 2p-method
    if row['h_2p'] > h_2p_max:
        row['h_2p'] = h_2p_max
    tmp.append(row)    
df = pandas.DataFrame(tmp)
del tmp
df['h_2p'].describe()
```

```python
df['th_2p']  = np.arcsin((df['gran']*1e-3)*np.cos(df['elv_init']*np.pi/180.) / (6370. + df['h_2p']))
df['lgc_2p'] = df['th_2p'] * 6370.
```

```python
df['nr_2p'] = np.zeros_like(df['th_2p'], dtype=np.float)
size = df['th_2p'].size
df['phase'] = np.zeros_like(df['beam'], dtype = np.int64)

print size



```

```python
df.sort(['elv_init', 'gran'], ascending=[False, True], inplace=True)
print size
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
initialize = 1
max_elv_init_gs = dgs['elv_init'][0]
for index in range(0, size-1):
    if initialize == 1:
        initialize = 0
        df['phase'][index] = 1
        inclination_init = np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)
        df['nr_2p'][index] = np.sin((90-df['elv_init'][index]-df['th_2p'][index]*180./np.pi)*np.pi/180.)/inclination_init
        last_elv = df['elv_init'][index]
        last_nr_2p = df['nr_2p'][index]
        last_gran = df['gran'][index]
        if index < size-1:
            df['phase'][index+1] = df['phase'][index]
        continue
        
    elif df['elv_init'][index] < last_elv and df['phase'][index] == 1:
        inclination_init = np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)
        df['nr_2p'][index] = np.sin((90-df['elv_init'][index]-df['th_2p'][index]*180./np.pi)*np.pi/180.)/inclination_init
#bks is 64.73 inc,  0.81/degree, sas is 75.3 inc, 0.6/degree
        if df['h_2p'][index] < 200.:
            if index < size-1:
                df['phase'][index+1] = df['phase'][index]
            continue
        last_elv = df['elv_init'][index]    
        last_nr_2p = df['nr_2p'][index]
        last_gran = df['gran'][index]
        if index < size-1:
            df['phase'][index+1] = df['phase'][index]
        continue    
    
    elif df['elv_init'][index] == last_elv and  df['phase'][index] == 1:
        df['nr_2p'][index] = last_nr_2p*inclination_init/np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)        
        if df['gran'][index] > last_gran + 50000:  # gran gap between bottomside and topeside scatter for penetraing rays.
            df['phase'][index] = 2
            last_gran = df['gran'][index]
            if index < size-1:
                df['phase'][index+1] = df['phase'][index]
            continue
        last_gran = df['gran'][index]
        if index < size-1:
            df['phase'][index+1] = df['phase'][index]
        continue
                
    elif df['elv_init'][index] < last_elv and df['phase'][index] == 2:
        if df['elv_init'][index] == max_elv_init_gs:
            last_index = index
            break
        inclination_init = np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)
        df['nr_2p'][index] = np.sin((90-df['elv_init'][index]-df['th_2p'][index]*180./np.pi)*np.pi/180.)/inclination_init
        if df['h_2p'][index] < 200.:
            if index < size-1:
                df['phase'][index+1] = df['phase'][index]
            continue
        last_elv = df['elv_init'][index]
        last_nr_2p = df['nr_2p'][index]
        if index < size-1:
            df['phase'][index+1] = df['phase'][index]
        continue
    
    elif df['elv_init'][index] == last_elv and df['phase'][index] == 2:
        df['nr_2p'][index] = last_nr_2p*inclination_init/np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)
        if index < size-1:
            df['phase'][index+1] = df['phase'][index]
        continue 
        
   
for index in range(last_index, size-1):
    df['phase'][index] = 3
    inclination_init = np.sin((64.7 + 0.8*df['lgc_2p'][index]/112.)*np.pi/180.)
    df['nr_2p'][index] = np.sin((90-df['elv_init'][index]-df['th_2p'][index]*180./np.pi)*np.pi/180.)/inclination_init
    continue     
     
df['nr_2p'][size-1] = df['nr_2p'][size-1] = np.sin((90-df['elv_init'][size-1]-df['th_2p'][size-1]*180./np.pi)*np.pi/180.)\
        /np.sin((64.7 + 0.8*df['lgc_2p'][size-1]/112.)*np.pi/180.)
df['phase'][size-1] = 3
     
```

```python
nr_offset = df['nr'].mean() - df['nr_2p'].mean()
print nr_offset
df['nr_2p'] = df['nr_2p'] + nr_offset
```
